```python
# Import necessary package
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
print(0)
```

```python
#Generating data 
#by Bernoulli distribution for binomial distribution and by the standard normal distribution for the other three distributions.[First]
data_type_1 = []

print('Generating data, please wait.')
data_type_1_generate = np.random.normal(size=(100000, 100))  #100000Group, 100numbers of each group
print('Successfully generated data!')

sum_normal = []
count = 0

print('Organizing data, please wait.')
for i in np.square(data_type_1_generate):
    sum_normal.append(np.sum(i))
    count = count + 1
    if (count%100) == 0:
        data_type_1.append(sum_normal)
        sum_normal = []

print('Successfully Organized data!')
print(0)
```

```python
#Generating data using certain statistical software [Second]
data_type_2 = []

print('Generating data, please wait.')
data_type_2 = np.random.chisquare(100, size=(1000,100)) # df=100, 1000 group, each group has 100 number
print('Successfully generated data!')
print(0)
```

```python
#Generating data by bootstrap, i.e., take a sample from the population and then resampling from the sample. [Third]
data_type_3 = []

print('Generating data, please wait.')
data_type_3 = random.sample(list(data_type_1), 1000)
print('Successfully generated data!')
print(0)
```

```python
def draw_graph(data):
    clean_data_mean = []
    clean_data_var = []
    clean_data_skew = []
    clean_data_skew_num = []
    clean_data_kurtosis = []
    clean_data_kurtosis_num = []
    clean_data_5_percentile = []
    clean_data_95_percentile = []

    print('Organizing data, please wait.')
    for d in data:
        clean_data_mean.append(np.mean(d))
        clean_data_var.append(np.var(d))
        clean_data_skew.append(pd.DataFrame(d).skew())
        clean_data_kurtosis.append(pd.DataFrame(d).kurtosis())
        clean_data_5_percentile.append(np.percentile(d, 5))
        clean_data_95_percentile.append(np.percentile(d, 95))

    for i in clean_data_skew:
        clean_data_skew_num.append(i[0])

    for i in clean_data_kurtosis:
        clean_data_kurtosis_num.append(i[0])
    print('Successfully organized the data!')

    print('Drawing, please wait.')

    #Histogram
    plt.figure(1)
    plt.style.use('Solarize_Light2')
    plt.hist(clean_data_mean,20,histtype='stepfilled',facecolor='r',alpha=0.5)
    plt.xlabel("mean",fontsize=13,fontweight='bold')
    plt.ylabel("frequency",fontsize=13,fontweight='bold')
    plt.title('Mean for chi-square')
    #plt.xticks(np.arange(1.4, 2.8, 0.2))

    plt.figure(2)
    plt.hist(clean_data_var,20,histtype='stepfilled',facecolor='b',alpha=0.5)
    plt.xlabel("variance",fontsize=13,fontweight='bold')
    plt.ylabel("frequency",fontsize=13,fontweight='bold')
    plt.title('Variance for chi-square')
    #plt.xticks(np.arange(1.4, 2.8, 0.2))

    plt.figure(3)
    plt.hist(clean_data_skew_num,20,histtype='stepfilled',facecolor='g',alpha=0.5)
    plt.xlabel("skewness",fontsize=13,fontweight='bold')
    plt.ylabel("frequency",fontsize=13,fontweight='bold')
    plt.title('Skewness for chi-square')
    #plt.xticks(np.arange(1.4, 2.8, 0.2))

    plt.figure(4)
    plt.hist(clean_data_kurtosis_num,20,histtype='stepfilled',facecolor='y',alpha=0.5)
    plt.xlabel("kurtosis",fontsize=13,fontweight='bold')
    plt.ylabel("frequency",fontsize=13,fontweight='bold')
    plt.title('Kurtosis for chi-square')
    #plt.xticks(np.arange(1.4, 2.8, 0.2))

    plt.figure(5)
    plt.hist(clean_data_5_percentile,20,histtype='stepfilled',facecolor='m',alpha=0.5)
    plt.xlabel("value",fontsize=13,fontweight='bold')
    plt.ylabel("frequency",fontsize=13,fontweight='bold')
    plt.title('5% quantitle for chi-square')
    #plt.xticks(np.arange(1.4, 2.8, 0.2))

    plt.figure(6)
    plt.hist(clean_data_95_percentile,20,histtype='stepfilled',facecolor='k',alpha=0.5)
    plt.xlabel("value",fontsize=13,fontweight='bold')
    plt.ylabel("frequency",fontsize=13,fontweight='bold')
    plt.title('95% quantitle for chi-square')
    #plt.xticks(np.arange(1.4, 2.8, 0.2))

    plt.show()

    #Box plot
    plt.figure(7)
    plt.title('Mean for chi-square')
    plt.grid(axis='y')
    bplot = plt.boxplot(clean_data_mean, patch_artist=True)
    colors = ['pink']
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

    plt.figure(8)
    plt.title('Variance for chi-square')
    plt.grid(axis='y')
    bplot = plt.boxplot(clean_data_var, patch_artist=True)
    colors = ['lightblue']
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

    plt.figure(9)
    plt.title('Skewness for chi-square')
    plt.grid(axis='y')
    bplot = plt.boxplot(clean_data_skew_num, patch_artist=True)
    colors = ['lightgreen']
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

    plt.figure(10)
    plt.title('Kurtosis for chi-square')
    plt.grid(axis='y')
    bplot = plt.boxplot(clean_data_kurtosis_num, patch_artist=True)
    colors = ['grey']
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

    plt.figure(11)
    plt.title('5% quantitle for chi-square')
    plt.grid(axis='y')
    bplot = plt.boxplot(clean_data_5_percentile, patch_artist=True)
    colors = ['lightyellow']
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

    plt.figure(12)
    plt.title('95% quantitle for chi-square')
    plt.grid(axis='y')
    bplot = plt.boxplot(clean_data_95_percentile, patch_artist=True)
    colors = ['orange']
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

    plt.show()

    #Q-Q plot
    plt.figure(13)
    sorted_ = np.sort(clean_data_mean)
    yvals = np.arange(len(sorted_))/float(len(sorted_))
    #plt.plot(sorted_, yvals)
    x_label = stats.norm.ppf(yvals)
    #Finding the inverse of the cumulative distribution function of
    #the standard normal distribution for the value of the target cumulative distribution function
    plt.scatter(x_label, sorted_)
    stats.probplot(clean_data_mean, dist="norm", plot=plt)
    plt.xlabel("Standard Normal Quantiles",fontsize=13,fontweight='bold')
    plt.ylabel("Quantiles of Input Sample",fontsize=13,fontweight='bold')
    plt.title('mean for chi-square')

    plt.figure(14)
    sorted_ = np.sort(clean_data_var)
    yvals = np.arange(len(sorted_))/float(len(sorted_))
    #plt.plot(sorted_, yvals)
    x_label = stats.norm.ppf(yvals)
    #Finding the inverse of the cumulative distribution function of
    #the standard normal distribution for the value of the target cumulative distribution function
    plt.scatter(x_label, sorted_)
    stats.probplot(clean_data_var, dist="norm", plot=plt)
    plt.xlabel("Standard Normal Quantiles",fontsize=13,fontweight='bold')
    plt.ylabel("Quantiles of Input Sample",fontsize=13,fontweight='bold')
    plt.title('Variance for chi-square')

    plt.figure(15)
    sorted_ = np.sort(clean_data_skew_num)
    yvals = np.arange(len(sorted_))/float(len(sorted_))
    #plt.plot(sorted_, yvals)
    x_label = stats.norm.ppf(yvals)
    #Finding the inverse of the cumulative distribution function of
    #the standard normal distribution for the value of the target cumulative distribution function
    plt.scatter(x_label, sorted_)
    stats.probplot(clean_data_skew_num, dist="norm", plot=plt)
    plt.xlabel("Standard Normal Quantiles",fontsize=13,fontweight='bold')
    plt.ylabel("Quantiles of Input Sample",fontsize=13,fontweight='bold')
    plt.title('Skewness for chi-square')

    plt.figure(16)
    sorted_ = np.sort(clean_data_kurtosis_num)
    yvals = np.arange(len(sorted_))/float(len(sorted_))
    #plt.plot(sorted_, yvals)
    x_label = stats.norm.ppf(yvals)
    #Finding the inverse of the cumulative distribution function of
    #the standard normal distribution for the value of the target cumulative distribution function
    plt.scatter(x_label, sorted_)
    stats.probplot(clean_data_kurtosis_num, dist="norm", plot=plt)
    plt.xlabel("Standard Normal Quantiles",fontsize=13,fontweight='bold')
    plt.ylabel("Quantiles of Input Sample",fontsize=13,fontweight='bold')
    plt.title('Kurtosis for chi-square')

    plt.figure(17)
    sorted_ = np.sort(clean_data_5_percentile)
    yvals = np.arange(len(sorted_))/float(len(sorted_))
    #plt.plot(sorted_, yvals)
    x_label = stats.norm.ppf(yvals)
    #Finding the inverse of the cumulative distribution function of
    #the standard normal distribution for the value of the target cumulative distribution function
    plt.scatter(x_label, sorted_)
    stats.probplot(clean_data_5_percentile, dist="norm", plot=plt)
    plt.xlabel("Standard Normal Quantiles",fontsize=13,fontweight='bold')
    plt.ylabel("Quantiles of Input Sample",fontsize=13,fontweight='bold')
    plt.title('5% quantitle for chi-square')

    plt.figure(18)
    sorted_ = np.sort(clean_data_95_percentile)
    yvals = np.arange(len(sorted_))/float(len(sorted_))
    #plt.plot(sorted_, yvals)
    x_label = stats.norm.ppf(yvals)
    #Finding the inverse of the cumulative distribution function of
    #the standard normal distribution for the value of the target cumulative distribution function
    #plt.scatter(x_label, sorted_, facecolor='y')
    plt.scatter(x_label, sorted_)
    stats.probplot(clean_data_95_percentile, dist="norm", plot=plt)
    plt.xlabel("Standard Normal Quantiles",fontsize=13,fontweight='bold')
    plt.ylabel("Quantiles of Input Sample",fontsize=13,fontweight='bold')
    plt.title('95% quantitle for chi-square')

    plt.show()
    print(0)
print(0)
```
